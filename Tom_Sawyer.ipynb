I have done the analysis only for the content of the book that is discarding header and footer texts from the file.



how do I select the text that I want to do the analysis only for?

1. spilt the text by white space that will give list of words including puncuations.
2. then removing punctuations using string.punctuation and translate function.
3.converting numbers into their word representation eg,   2->two
4. selecting the content by getting index of words where content starts and ends and then slicing the  word list using that indices.


```
# split into words by white space

words = text.split()

# remove punctuation from each word

table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words] # stripped -> without punctuation but includes numbers

# convert numbers to their word representation 2-> two
# import the inflect library 
import inflect 
p = inflect.engine()
word_list = []  # initializing a list:  word_list -> numbers in their word representation 
  
for word in stripped: 
    # if word is a digit, convert the digit 
    # to numbers and append into the word_list list 
    if word.isdigit(): 
        temp = p.number_to_words(word) 
        word_list.append(temp) 

    # append the word as it is 
    else: 
        word_list.append(word) 
print(word_list.index("HARTFORD")) ## knowing the index where the content starts
print(word_list[1089:])
```

```
word_list.index("CONCLUSION") #getting the index of end of the content
```

case-1: Analysis at word level

```
dicts = {}
for i in word_type:
        dicts[i] = content.count(i)
print(dicts)
```

```
sorted_dict = {k: v for k, v in sorted(dicts.items(), key=lambda item: item[1],reverse=True)} 
```

```
sorted_dict
```

```
# changing keys of dictionary to rank value rather than word-type value
x=range(1,len(word_type)+1) # rank
rank_dict = dict(zip(x, list(sorted_dict.values()))) 
rank_dict
```

```
import csv
w = csv.writer(open("/content/freq.csv", "w"))
for key, val in rank_dict.items():
    w.writerow([key, val])
```

```
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r'/content/freq.csv',names=["rank", "frequency"])
data.head()
```

```
plt.scatter(data['rank'],data['frequency'])
plt.xlabel('rank')
plt.ylabel('frequency')
plt.title('rank-frequency plot')
```

```
# Create figure 
fig, ax = plt.subplots() 
  
# log x and y axis 
ax.loglog(data['rank'],data['frequency']) 
plt.xlabel('log-rank')
plt.ylabel('log-frequency') 
ax.set_title('Log-Rank-Log-frequency plot') 
plt.show()
```

```
from scipy.stats import pearsonr
# calculate Pearson's correlation
corr, _ = pearsonr(data['rank'],data['frequency'])
print('Pearsons correlation for words: %.3f' % corr)
```

This strong negative correlation signifies that as the word become more frequent(high frequency), the rank of the word decreases(low rank) (and vice versa)
